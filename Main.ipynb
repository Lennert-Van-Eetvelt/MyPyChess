```python
import PySimpleGUI as sg

CHESS_PATH = "sprites"

blank = os.path.join(CHESS_PATH, 'blank.png')
bishopB = os.path.join(CHESS_PATH, 'bishopb.png')
bishopW = os.path.join(CHESS_PATH, 'bishopw.png')
pawnB = os.path.join(CHESS_PATH, 'pawnb.png')
pawnW = os.path.join(CHESS_PATH, 'pawnw.png')
knightB = os.path.join(CHESS_PATH, 'knightb.png')
knightW = os.path.join(CHESS_PATH, 'knightw.png')
rookB = os.path.join(CHESS_PATH, 'rookb.png')
rookW = os.path.join(CHESS_PATH, 'rookw.png')
queenB = os.path.join(CHESS_PATH, 'queenb.png')
queenW = os.path.join(CHESS_PATH, 'queenw.png')
kingB = os.path.join(CHESS_PATH, 'kingb.png')
kingW = os.path.join(CHESS_PATH, 'kingw.png')

images = {'BishopB': bishopB, 'BishopW': bishopW, 'PawnB': pawnB, 'PawnW': pawnW, 'KnightB': knightB, 'KnightW': knightW,
          'RookB': rookB, 'RookW': rookW, 'KingB': kingB, 'KingW': kingW, 'QueenB': queenB, 'QueenW': queenW, 'BLANK': blank}



def getLayout():
    layout = [[sg.T('', pad= (80,0))] + [sg.T('{}'.format(a), pad=((23,27),0), font='Any 13') for a in 'abcdefgh']]
    for y in range(board.size):
        row = []
        row.append(sg.RButton("",image_filename=images.get("BLANK"),  size=(1, 1), pad = (0,0), button_color=('white', 'gray'), key="White"+str(y)))
        row.append(sg.RButton("",image_filename=images.get("BLANK"),  size=(1, 1), pad = ((0,0),0), button_color=('white', 'gray'), key="White"+str(y   +board.size)))
        row.append(sg.T(str(board.size-y)+'   ', font='Any 13'))
        for x in range(board.size):
            color = 'white' if (x+y) % 2 else 'lightslategray'
            piece = board.get(x,y)
            pieceImage = "BLANK" if piece is None else piece.name + piece.color[0] 
            row.append(sg.RButton("",image_filename=images.get(pieceImage),  size=(1, 1), pad = (0,0), button_color=('white', color),key = (x,y)))
        row.append(sg.T(str(board.size-y)+'   ', font='Any 13'))
        row.append(sg.RButton("",image_filename=images.get("BLANK"),  size=(1, 1), pad = ((0,0),0), button_color=('white', 'gray'), key="Black"+str(y)))
        row.append(sg.RButton("",image_filename=images.get("BLANK"),  size=(1, 1), pad = (0,0), button_color=('white', 'gray'), key="Black"+str(y   +board.size)))
        layout.append(row)
    layout.append([sg.T('',pad= (80,0))] + [sg.T('{}'.format(a), pad=((23,27),0), font='Any 13') for a in 'abcdefgh'])
    return layout


def redraw_board(window, board, moves, selected, state):
    for y in range(board.size):
        for x in range(board.size):
            color = 'white' if (x+y) % 2 else 'lightslategray'
            for move in moves:
                if move == (x,y): 
                    color = 'magenta' if color is 'white' else 'purple'
            piece = board.get(x,y)
            if selected == (x,y) or (state is "promotion" and piece is not None and (piece.name is "Rook"or piece.name is "Queen" or piece.name is "Bishop" or piece.name is "Knight")):
                color = 'lightgreen' if (x+y) % 2 else 'lime green'
            if state is "promotion" and selected == (x,y):
                color = 'dodgerblue2' if (x+y) %2 else 'dodgerblue3'
            pieceImage = "BLANK" if piece is None else piece.name + piece.color[0] 
            elem = window.FindElement(key=(x,y))
            elem.Update(button_color = ('white', color),
                        image_filename=images.get(pieceImage),)
    whiteCounter = 0
    blackCounter = 0
    for piece in board.dead:
        counter = whiteCounter if piece.color is 'White' else blackCounter
        elem = window.FindElement(key = piece.color + str(counter))
        color = "gray"
        if state is "promotion" and piece is not None and (piece.name is "Rook"or piece.name is "Queen" or piece.name is "Bishop" or piece.name is "Knight"):
            color = 'lightgreen'
        elem.Update(image_filename= images.get(piece.name + piece.color[0]), button_color=('white', color))
        if piece.color == "White":
            whiteCounter += 1
        else:
            blackCounter += 1



window = sg.Window("Chess", getLayout(), resizable=True)

selPos = (-1,-1)
sel = False
state = ""
while True:
    button, values = window.read()
    if button is None or button == sg.WIN_CLOSED:
        break
    if type(button) is tuple:
        moved = False
        selected = (-1,-1)
        moves = []
        piece = board.get(*button)
        if sel:
            for move in  board.get(*selPos).possibleMoves():
                if move == button:
                    board.get(*selPos).move(*button)
                    player = 'Black' if player is 'White' else 'White'
                    moved = True
            sel = False
        if not moved and piece is not None and piece.color == player:
            sel = True
            selPos = button
            moves = piece.possibleMoves()
        if sel:
            selected = selPos
        if board.getState() =="promotion":
            while True:
                redraw_board(window,board,[],button,board.getState())
                button2, values2 = window.read()
                if button2 == sg.WIN_CLOSED:
                    button = sg.WIN_CLOSED
                    print('close')
                    break
                if type(button2) is tuple:
                    newPiece = board.get(*button2)
                    if newPiece is not None and board.get(*button).promote(newPiece):
                        board.cleanState()
                        selected = (-1,-1)
                        break
                if type(button2) is str:
                    if button2.startswith("Black") or button2.startswith('White'):
                        deadnumber = int(button2[5:])
                        deadcolor = button2[:5]
                        k = 0
                        for deadpiece in board.dead:
                            if deadpiece.color == deadcolor:
                                if k == deadnumber:
                                    board.get(*button).promote(deadpiece)
                                    board.cleanState()
                                    selected = (-1,-1)
                                    break
                                k +=1
                        break
        redraw_board(window, board, moves, selected, board.getState())
        board.cleanState()
        board.isCheckmate(player)
window.close()
```
